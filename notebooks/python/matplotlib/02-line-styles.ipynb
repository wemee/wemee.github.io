# 02 · 把一條線畫到好看：顏色、線型、標記、圖例

上一課你學會了 `fig, ax = plt.subplots()` 的核心節奏。這課我們專心做一件事：**把一條（或幾條）折線畫到清楚又好看**。

> 💡 這是可執行的 notebook，每個格子按 `Shift+Enter` 就會跑。大膽改顏色、改數字。

## 學習目標

- 控制線的 **顏色 (color)**、**粗細 (linewidth)**、**線型 (linestyle)**、**標記 (marker)**
- 一張圖畫**多條線**，並用 **圖例 (legend)** 區分
- 加上 **格線 (grid)** 與軸標籤，讓圖能被讀懂

## 1. 從預設到客製

先畫一條最樸素的線，再一個參數一個參數加上去，你就能看清每個參數在做什麼。

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 30)
y = np.sin(x)

fig, ax = plt.subplots()

ax.plot(
    x, y,
    color="crimson",     # 顏色
    linewidth=2.5,        # 線粗
    linestyle="--",      # 線型：'-' 實線, '--' 虛線, ':' 點線, '-.' 點劃線
    marker="o",          # 資料點標記：'o' 圓, 's' 方, '^' 三角
    markersize=6,
)
ax.set_title("One styled line")
plt.show()
```

**動手試試**：把 `linestyle` 改成 `":"`、`marker` 改成 `"^"`、`color` 改成 `"steelblue"`，跑跑看差別。

顏色可以這樣寫：
- 名稱：`"red"`、`"crimson"`、`"steelblue"`
- 十六進位：`"#1f77b4"`
- matplotlib 預設色盤：`"tab:blue"`、`"tab:orange"`、`"tab:green"`…（最推薦，整組搭起來好看）

## 2. 多條線 + 圖例

每條線給一個 `label`，最後呼叫 `ax.legend()`，matplotlib 就會自動把圖例畫出來。

```python
x = np.linspace(0, 2 * np.pi, 200)

fig, ax = plt.subplots(figsize=(7, 4))

ax.plot(x, np.sin(x), color="tab:blue", label="sin(x)")
ax.plot(x, np.cos(x), color="tab:orange", label="cos(x)")
ax.plot(x, np.sin(x) * np.cos(x), color="tab:green", linestyle="--", label="sin·cos")

ax.set_title("Three curves")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()                 # 自動依 label 產生圖例
ax.grid(True, alpha=0.3)    # 淡淡的格線，alpha 控制透明度
plt.show()
```

`ax.legend()` 還可以指定位置，例如 `ax.legend(loc="upper right")` 或 `loc="best"`（讓 matplotlib 自己挑空位）。

## 常用參數速查

| 參數 | 作用 | 常見值 |
| --- | --- | --- |
| `color` | 顏色 | `"tab:blue"`、`"#1f77b4"`、`"crimson"` |
| `linewidth` / `lw` | 線粗 | `1`、`2.5` |
| `linestyle` / `ls` | 線型 | `"-"`、`"--"`、`":"`、`"-."` |
| `marker` | 資料點記號 | `"o"`、`"s"`、`"^"`、`"."` |
| `alpha` | 透明度 | `0`（全透明）～`1`（不透明） |
| `label` | 圖例文字 | 任意字串，搭配 `ax.legend()` |

## 小結

- 線的外觀靠 `color` / `linewidth` / `linestyle` / `marker` 四個參數組合。
- 多條線：每條給 `label`，再 `ax.legend()`。
- `ax.grid(True, alpha=0.3)` 加淡格線幫助讀值。
