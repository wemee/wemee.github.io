# 03 · 圖型動物園：散佈、長條、直方、箱型

折線圖只是開始。這課帶你認識四種最常用的圖，重點不是「怎麼畫」，而是**什麼資料該用哪一種**——選錯圖型，再漂亮也會誤導人。

> 💡 可執行 notebook，`Shift+Enter` 跑每一格。

## 學習目標

- `scatter`（散佈圖）看兩個變數的**關係**
- `bar`（長條圖）比較**類別**之間的數量
- `hist`（直方圖）看單一變數的**分布**
- `boxplot`（箱型圖）看資料的**離散程度與離群值**
- 建立「看到資料就知道選哪種圖」的直覺

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.default_rng(42)   # 固定亂數種子，結果可重現
```

## 1. scatter — 兩個變數有沒有關係？

每個點是一筆資料的 (x, y)。適合問：「身高和體重有關嗎？」「廣告花費和銷售額呢？」

```python
height = rng.normal(170, 8, 80)
weight = 0.9 * height - 90 + rng.normal(0, 5, 80)   # 與身高正相關

fig, ax = plt.subplots()
ax.scatter(height, weight, alpha=0.6, color="tab:blue")
ax.set_xlabel("Height (cm)")
ax.set_ylabel("Weight (kg)")
ax.set_title("Scatter: relationship")
plt.show()
```

## 2. bar — 比較類別

每根長條代表一個類別的數量。注意 x 軸是**類別**（不連續），這是和直方圖最大的差別。

```python
langs = ["Python", "JS", "Go", "Rust", "C++"]
stars = [42, 38, 21, 18, 25]

fig, ax = plt.subplots()
ax.bar(langs, stars, color="tab:green")
ax.set_ylabel("Stars (k)")
ax.set_title("Bar: compare categories")
plt.show()
```

## 3. hist — 單一變數的分布

把資料切成一格一格的「bin」，數每格有幾筆。適合問：「考試成績集中在哪？」「大部分使用者停留幾秒？」

```python
scores = rng.normal(72, 12, 500)

fig, ax = plt.subplots()
ax.hist(scores, bins=20, color="tab:orange", edgecolor="white")
ax.set_xlabel("Score")
ax.set_ylabel("Count")
ax.set_title("Histogram: distribution")
plt.show()
```

**動手試試**：把 `bins` 從 20 改成 5、再改成 50，感受 bin 數量怎麼影響你對分布的判讀。

## 4. boxplot — 離散程度與離群值

箱子是中間 50% 的資料，中線是中位數，鬚線外的點是離群值。最適合**並排比較幾組資料的分散狀況**。

```python
group_a = rng.normal(60, 10, 100)
group_b = rng.normal(75, 5, 100)
group_c = rng.normal(70, 20, 100)

fig, ax = plt.subplots()
ax.boxplot([group_a, group_b, group_c], tick_labels=["A", "B", "C"])
ax.set_ylabel("Value")
ax.set_title("Boxplot: spread & outliers")
plt.show()
```

## 該用哪種圖？決策表

| 你想問的問題 | 用這個 |
| --- | --- |
| 兩個數值變數有沒有關係 | `scatter` |
| 不同類別誰多誰少 | `bar` |
| 一個變數怎麼分布 | `hist` |
| 幾組資料的分散/離群比較 | `boxplot` |
| 隨時間的變化趨勢 | `plot`（折線，第 02 課） |

> ⚠️ 關於圓餅圖 `pie`：能少用就少用。人眼比較「角度」遠不如比較「長度」，同樣的資料用長條圖幾乎都更清楚。

## 小結

- 選圖型的關鍵是**你要回答什麼問題**，不是哪個好看。
- `bar`（類別）vs `hist`（連續分布）最容易搞混，記住 x 軸性質不同。
